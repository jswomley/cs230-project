```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
model_path = "./saved_model"
model = tf.saved_model.load(model_path)
```

```python
image = tf.keras.preprocessing.image.load_img("../test_image_2.png", target_size=(256,256), keep_aspect_ratio=True)
input_arr = tf.keras.preprocessing.image.img_to_array(image)
plt.imshow(image)
plt.show()
input_arr = (input_arr / 127.5) - 1
input_arr = tf.convert_to_tensor([input_arr])
predictions = model(input_arr, training=True)
```

```python
output_image = predictions[0]
plt.imshow(output_image)
plt.colorbar()
```
